In a new code cell, create and show a figure that includes one axis with a bar plot of the `Evaluation` variable

```python
eval_freq = cces.groupby("Evaluation").size()
eval_freq
```

```python
fig1, ax1 = plt.subplots()
x = eval_freq.index
y = eval_freq.values
ax1.bar(x, y)
ax1.set_xlabel("Evaluation")
ax1.set_ylabel("Frequency")
plt.show()
```

In a new code cell, create and show a figure that includes one axis with a scatter plot of the `ConMediaUse` and `ImmiPhobia` variable

```python
fig2, ax2 = plt.subplots()
ax2.scatter(cces["ImmiPhobia"], cces["age"], color = "yellow", label ="ImmiPhobia")
ax2.scatter(cces["ConMediaUse"],cces["age"], color = "blue", label ="ConMediaUse")
ax2.set_xlabel("ConMediaUse and ImmiPhobia")
ax2.set_ylabel("Age")
plt.show()
```
